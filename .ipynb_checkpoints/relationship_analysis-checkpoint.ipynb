```python
import pandas as pd
import numpy as np
import spacy
from spacy import displacy
import networkx as nx

import matplotlib.pyplot as plt
```

```python
# Load spacy English language Model
NER = spacy.load("en_core_web_sm")
from spacy.lang.en import English
```

## Load Book

```python
import os
 
# Get all the books in the data directory
all_books = [b for b in os.scandir('text/Dune') if '.txt' in b.name]
```

```python
all_books
```

```python
book = all_books[1]
book_text = open(book).read()
book_doc = NER(book_text)
```

```python
# Visualize identified entities
displacy.render(book_doc[0:2000], style="ent", jupyter=True)
```

## Load Character Names

```python
# Read in characters
character_df = pd.read_csv("text/Dune/characters.csv")
```

```python
character_df.head()
```

```python
# Remove parentheticals and text within 
import re 
character_df['character'] = character_df['character'].apply(lambda x: re.sub("[\(].*?[\)]", "", x))

# Get character's first name
character_df['character_firstname'] = character_df['character'].apply(lambda x: x.split(' ', 1)[0])
```

```python
pd.set_option('display.max_rows', None)
character_df
```

## Get Named Entity List per Sentence

```python
sent_entity_df = []

# Loop through sentences and store named entity for each sentence
for sent in book_doc.sents:
    entity_list = [ent.text for ent in sent.ents]
    sent_entity_df.append({"sentence": sent, 
                          "entities": entity_list})

sent_entity_df = pd.DataFrame(sent_entity_df)
```

```python
sent_entity_df
```

```python
# Function to filter out non-character entities
def filter_entity(ent_list, character_df):
    return [ent for ent in ent_list
           if ent in list(character_df.character)
           or ent in list(character_df.character_firstname)]
```

```python
sent_entity_df['character_entities'] = sent_entity_df['entities'].apply(lambda x: filter_entity(x, character_df))

# Filter out sentences that have no character entities
sent_entity_df_filtered = sent_entity_df[sent_entity_df['character_entities'].map(len) > 0]
sent_entity_df_filtered.head(10)
```

```python
# Take the first names of characters
sent_entity_df_filtered['character_entities'] = sent_entity_df_filtered['character_entities'].apply(lambda x: [item.split()[0]
                                                                                                                for item in x])
```

```python
pd.reset_option('^display.', silent=True)
sent_entity_df_filtered
```

## Create Relationships

```python
window_size = 5
relationships = []

for i in range(sent_entity_df_filtered.index[-1]):
    end_i = min(i+5, sent_entity_df_filtered.index[-1])
    char_list = sum((sent_entity_df_filtered.loc[i: end_i].character_entities), [])
    
    # Remove duplicate characters next to each other
    char_unique = [char_list[i] for i in range(len(char_list))
                  if (i==0) or char_list[i] != char_list[i-1]]
    
    if len(char_unique) > 1:
        for idx, a in enumerate(char_unique[:-1]):
            b = char_unique[idx + 1]
            relationships.append({"source": a, "target": b})
```

```python
relationship_df = pd.DataFrame(relationships)
```

```python
pd.set_option('display.max_rows', None)
relationship_df
```

```python
# Sort the cases with a -> b and b -> a
relationship_df = pd.DataFrame(np.sort(relationship_df.values, axis = 1), columns = relationship_df.columns)
relationship_df
```

```python
relationship_df["value"] = 1
relationship_df = relationship_df.groupby(["source", "target"], sort=False, as_index=False).sum()
```

```python
relationship_df.head(10)
```

## Graph Analysis and Visualization

```python
# Create a graph from a pandas dataframe
G = nx.from_pandas_edgelist(relationship_df, 
                           source = "source",
                           target = "target",
                           edge_attr = "value",
                           create_using = nx.Graph())
```

### Graph Visualization - Networkx

```python
plt.figure(figsize=(10,10))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, with_labels=True, node_color='green', edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```

### Graph Visualization - Pyvis

```python
from pyvis.network import Network
net = Network(notebook = True, width="1000px", height="700px", bgcolor="#222222", font_color="white")

node_degree = dict(G.degree)

# Setting up node size attribute
nx.set_node_attributes(G, node_degree, 'size')

net.from_nx(G)
net.show("dune.html")
```

### Most Important Characters in Dune

```python
# Degree centrality
degree_dict = nx.degree_centrality(G)
degree_dict
```

```python
degree_df = pd.DataFrame.from_dict(degree_dict, orient='index', columns=['centrality'])

# Plot top 10 nodes
degree_df.sort_values('centrality', ascending=False)[0:9].plot(kind="bar")
```

```python
# Betweenness centrality
betweenness_dict = nx.betweenness_centrality(G)
betweenness_df = pd.DataFrame.from_dict(betweenness_dict, orient='index', columns=['centrality'])

# Plot top 10 nodes
betweenness_df.sort_values('centrality', ascending=False)[0:9].plot(kind="bar")
```

```python
# Closeness centrality
closeness_dict = nx.closeness_centrality(G)
closeness_df = pd.DataFrame.from_dict(closeness_dict, orient='index', columns=['centrality'])

# Plot top 10 nodes
closeness_df.sort_values('centrality', ascending=False)[0:9].plot(kind="bar")
```

```python
# Save centrality measures
nx.set_node_attributes(G, degree_dict, 'degree_centrality')
nx.set_node_attributes(G, betweenness_dict, 'betweenness_centrality')
nx.set_node_attributes(G, closeness_dict, 'closeness_centrality')
```

### Evolution of Character's Importance

```python
# Initialize empty list for graphs from books
books_graph = []

# Loop through book list and create graphs
for book in all_books:
    book_text = open(book).read()
    book_doc = NER(book_text)
    
    # Get list of entities per sentences
    sent_entity_df = []

        # Loop through sentences, store named entity list for each sentence
    for sent in book_doc.sents:
        entity_list = [ent.text for ent in sent.ents]
        sent_entity_df.append({"sentence": sent, "entities": entity_list})
    
    sent_entity_df = pd.DataFrame(sent_entity_df)
    
    # Select only character entities
    sent_entity_df['character_entities'] = sent_entity_df['entities'].apply(lambda x: filter_entity(x, character_df))

    # Filter out sentences that don't have any character entities
    sent_entity_df_filtered = sent_entity_df[sent_entity_df['character_entities'].map(len) > 0]
    
    # Take only first name of characters
    sent_entity_df_filtered['character_entities'] = sent_entity_df_filtered['character_entities'].apply(lambda x: [item.split()[0] 
                                                                                                               for item in x])

    # Create relationship df
    window_size = 5
    relationships = []

    for i in range(sent_entity_df_filtered.index[-1]):
        end_i = min(i+5, sent_entity_df_filtered.index[-1])
        char_list = sum((sent_entity_df_filtered.loc[i: end_i].character_entities), [])
    
        # Remove duplicate characters next to each other
        char_unique = [char_list[i] for i in range(len(char_list))
                      if (i==0) or char_list[i] != char_list[i-1]]
    
        if len(char_unique) > 1:
            for idx, a in enumerate(char_unique[:-1]):
                b = char_unique[idx + 1]
                relationships.append({"source": a, "target": b})                                                                                                           
    
    relationship_df = pd.DataFrame(relationships)
    
    # Sort the cases with a -> b and b -> a
    relationship_df = pd.DataFrame(np.sort(relationship_df.values, axis = 1), columns = relationship_df.columns)
    
    relationship_df["value"] = 1
    relationship_df = relationship_df.groupby(["source", "target"], sort=False, as_index=False).sum()

    # Create a graph from a pandas dataframe
    G = nx.from_pandas_edgelist(relationship_df, 
                                source = "source", 
                                target = "target", 
                                edge_attr = "value", 
                                create_using = nx.Graph())     
    
    books_graph.append(G) 
```

```python
# Creating a list of degree centrality of all the books
evol = [nx.degree_centrality(book) for book in books_graph]

# Creating a DataFrame from the list of degree centralities in all the books
degree_evol_df = pd.DataFrame.from_records(evol)

# Plotting the degree centrality evolution of 5 main characters
degree_evol_df[["Duncan", "Jessica", "Alia", "Chani", "Leto"]].plot()
```
